```python
import numpy as np
import pandas as pd 
import matplotlib.pyplot as plt 
import seaborn as sns
```

```python
drug_df=pd.read_csv('E:\Python_Library\Machine_learning\DATASET\drug200.csv')
drug_df.head()
```

```python
drug_df.duplicated().sum()
```

```python
drug_df.isnull().sum()
```

```python
drug_df['Sex'].unique()
dex_dt = {'F':1,'M':0}

drug_df['Sex']=drug_df['Sex'].map(dex_dt)
drug_df['Drug'].unique()
dru_dt = {'drugY':0, 'drugC':1, 'drugX':2, 'drugA':3, 'drugB':4}
drug_df['Drug'] = drug_df['Drug'].map(dru_dt) 
drug_df['BP'].unique()
Bp_dt = {'HIGH':1,'LOW':0,'NORMAL':10}
drug_df['BP']=drug_df['BP'].map(Bp_dt)
drug_df['Cholesterol'].unique()
chol_dt = {'HIGH':1,'NORMAL':0}
drug_df['Cholesterol'] = drug_df['Cholesterol'].map(chol_dt) 
```

```python
drug_df.info()
```

```python
matrix=drug_df.corr()
plt.figure(figsize=(4,4))
sns.heatmap(matrix,annot=True)
plt.show()
```

```python
sns.scatterplot(x=drug_df['Age'],y=drug_df['Drug'])
plt.show()
```

```python
sns.relplot(x='Drug',y='Age',data=drug_df,kind='scatter',hue='BP',col='Drug')
plt.show()
```

```python
x =drug_df.drop('Drug',axis=1)
y = drug_df[['Drug']]
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test ,y_train,y_test = train_test_split(x,y,test_size=0.2,random_state=42)
x_train

```

```python
from sklearn.tree import DecisionTreeClassifier
dtc = DecisionTreeClassifier()
```

```python
dtc.fit(x_train,y_train)
```

```python
print(dtc.score(x_train,y_train))
```

```python
print(dtc.score(x_test,y_test))
```
